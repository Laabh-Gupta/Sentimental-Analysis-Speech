
# Speech Emotion Recognition — PyTorch Baselines (Multi‑Model + Visualizations)

This notebook implements a **complete, extensible** Speech Emotion Recognition (SER) pipeline in PyTorch, with:
- Multiple models: **CNN2D (mel-spec)**, **CRNN (mel-spec + BiLSTM)**, **CNN1D (waveform)**.
- On-the-fly feature extraction (torchaudio) for **log-mel** and **MFCC**.
- Robust dataset discovery under `./archive` (supports common SER datasets like RAVDESS, CREMA-D, TESS by filename/folder patterns).
- **Training utilities**: mixed precision (AMP), early stopping, checkpoints.
- **Visualizations**: waveforms, spectrograms, **training curves**, **confusion matrices**, and **UMAP** embeddings.
- **Comparison table** across models with accuracy/F1 and per-class metrics.

> **Data path:** This expects your audio dataset(s) under `./archive`. You can keep multiple datasets inside that folder; labels will be inferred from folder names or common filename patterns.

```python
import os, re, math, json, random, time
from pathlib import Path
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR
import torchaudio
import torchaudio.transforms as T
import librosa
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from sklearn.preprocessing import LabelEncoder
from tqdm.auto import tqdm

# Optional embeddings visualization
try:
    import umap
    HAS_UMAP = True
except Exception:
    HAS_UMAP = False

# ---- Config ----
DATA_ROOT = Path('archive')   # <----- your dataset root
SR = 16000                    # target sample rate
NUM_MELS = 64
N_FFT = 1024
HOP_LENGTH = 256
WIN_LENGTH = 1024
FMIN = 20
FMAX = 7600
FIXED_SEC = 4.0               # pad/trim audio to this duration
BATCH_SIZE = 8
EPOCHS = 35
LR = 1e-3
WEIGHT_DECAY = 1e-4
NUM_WORKERS = 0
SEED = 42
AMP = True                    # mixed precision training

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
DEVICE

print(DEVICE)
```

```python
# RAVDESS emotion map
RAVDESS_EMO = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised",
}

def infer_label_from_path(p: Path):
    """Infer emotion label from RAVDESS/CREMA/TESS-like paths."""
    name = p.name
    parent = p.parent.name.lower()

    # 1) TESS/ folder-based datasets
    folder_label = re.sub(r'[^a-z]+', '', parent)
    if folder_label in ['angry','disgust','fearful','happy','neutral','sad','surprised','calm','boredom']:
        return folder_label

    # 2) RAVDESS: emotion is the **3rd** code in the filename, tokens[2]
    # Example: 03-01-05-01-02-01-12.wav -> tokens = ['03','01','05',...]; '05'-> angry
    tokens = re.split(r'[-_\.]', name)
    if len(tokens) >= 3 and tokens[2].isdigit() and len(tokens[2]) == 2:
        code = tokens[2]
        if code in RAVDESS_EMO:
            return RAVDESS_EMO[code]

    # 3) CREMA-D: _ANG_, _DIS_, ...
    for k in EMO_KEYS_CREMA:
        if k in name:
            return EMO_MAP_CREMA[k]

    # 4) Fallback keywords
    lname = name.lower()
    for token in ['angry','disgust','fearful','happy','neutral','sad','surprised','calm','boredom']:
        if token in lname:
            return token

    return None
```

```python

from glob import glob

audio_exts = ('*.wav','*.flac','*.mp3','*.m4a','*.ogg')
files = []
root = (DATA_ROOT / 'audio_speech_actors_01-24') if (DATA_ROOT / 'audio_speech_actors_01-24').exists() else DATA_ROOT
for ext in audio_exts:
    files += list(root.rglob(ext))


print(f"Found {len(files)} audio files under {DATA_ROOT.resolve()}");

rows = []
for p in files:
    label = infer_label_from_path(p)
    if label is None:
        continue
    rows.append({'path': str(p), 'label': label})

df = pd.DataFrame(rows)
if len(df)==0:
    raise RuntimeError("No labeled files inferred. Adjust infer_label_from_path or dataset layout.")

print(df['label'].value_counts())
df.head()
```

```python

le = LabelEncoder()
df['y'] = le.fit_transform(df['label'])

# Speaker-independent split
df["actor"] = df["path"].apply(lambda p: int(Path(p).stem.split("-")[-1]))

train_actors = list(range(1, 17))   # Actors 1–16
val_actors   = list(range(17, 21))  # Actors 17–20
test_actors  = list(range(21, 25))  # Actors 21–24

train_df = df[df["actor"].isin(train_actors)]
val_df   = df[df["actor"].isin(val_actors)]
test_df  = df[df["actor"].isin(test_actors)]

CLASSES = list(le.classes_)
NUM_CLASSES = len(CLASSES)
print('Classes:', CLASSES)
print('train/val/test:', len(train_df), len(val_df), len(test_df))
```

```python

def load_wave(path, target_sr=SR, fixed_sec=FIXED_SEC):
    wav, sr = torchaudio.load(path)
    if wav.shape[0] > 1:
        wav = wav.mean(0, keepdim=True)  # mono
    if sr != target_sr:
        wav = T.Resample(sr, target_sr)(wav)
    # pad/trim to fixed length
    num_target = int(target_sr * fixed_sec)
    if wav.shape[-1] > num_target:
        wav = wav[:, :num_target]
    elif wav.shape[-1] < num_target:
        pad = num_target - wav.shape[-1]
        wav = F.pad(wav, (0, pad))
    return wav.squeeze(0)  # [T]

mel_transform = T.MelSpectrogram(
    sample_rate=SR, n_fft=N_FFT, hop_length=HOP_LENGTH, win_length=WIN_LENGTH,
    f_min=FMIN, f_max=FMAX, n_mels=NUM_MELS, power=2.0
)
amplitude_to_db = T.AmplitudeToDB()

mfcc_transform = T.MFCC(
    sample_rate=SR, n_mfcc=40, melkwargs={
        'n_fft': N_FFT, 'n_mels': NUM_MELS, 'hop_length': HOP_LENGTH, 'f_min': FMIN, 'f_max': FMAX
    }
)

class SERDataset(Dataset):
    def __init__(self, frame: pd.DataFrame, feature='mel', augment=False):
        assert feature in ('mel', 'mfcc', 'wave'), f"Bad feature={feature}"
        self.frame = frame.reset_index(drop=True)
        self.feature = feature
        self.augment = augment


    def __len__(self): return len(self.frame)

    def __getitem__(self, idx):
        row = self.frame.iloc[idx]
        wav = load_wave(row['path'])  # [T]
    
        # --- light waveform augmentation (independent of feature type) ---
        if self.augment:
            # gain jitter
            wav = wav * (0.8 + 0.4 * torch.rand(1))
    
            # noise
            if random.random() < 0.3:
                wav = wav + 0.005 * torch.randn_like(wav)
    
            # pitch shift (librosa API expects kwargs)
            if random.random() < 0.3:
                w = wav.numpy().astype(np.float32)
                w = librosa.effects.pitch_shift(y=w, sr=SR, n_steps=random.uniform(-2, 2))
                wav = torch.tensor(w, dtype=torch.float32)
    
        # --- feature extraction ---
        if self.feature == 'mel':
            spec = mel_transform(wav)
            spec = amplitude_to_db(spec).clamp(min=-80.0)       # [n_mels, time]
            x = (spec - spec.mean()) / (spec.std() + 1e-6)
            x = x.unsqueeze(0)                                  # [1, n_mels, time]
    
            # simple SpecAugment (only when augment=True)
            if self.augment:
                t = x.shape[-1]
                if t > 16:
                    t0 = random.randint(0, t - 16)
                    x[:, :, t0:t0+16] = 0.0
                f = x.shape[1]
                if f > 8:
                    f0 = random.randint(0, f - 8)
                    x[:, f0:f0+8, :] = 0.0
    
        elif self.feature == 'mfcc':
            mf = mfcc_transform(wav)                            # [n_mfcc, time]
            x = (mf - mf.mean())/(mf.std()+1e-6)
            x = x.unsqueeze(0)                                  # [1, n_mfcc, time]
    
        elif self.feature == 'wave':
            x = wav.unsqueeze(0)                                # [1, T]
    
        else:
            raise ValueError('Unknown feature type')
    
        y = int(row['y'])
        return x, y, row['path']

```

```python

def make_loaders(feature='mel', augment=False):
    PIN = (DEVICE.type == 'cuda')  # ✅ this line is valid here

    tr = SERDataset(train_df, feature=feature, augment=augment)
    va = SERDataset(val_df,   feature=feature, augment=False)
    te = SERDataset(test_df,  feature=feature, augment=False)

    return (
        DataLoader(tr, batch_size=BATCH_SIZE, shuffle=True,  num_workers=NUM_WORKERS, pin_memory=PIN),
        DataLoader(va, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=PIN),
        DataLoader(te, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=PIN),
    )


train_loader_mel, val_loader_mel, test_loader_mel = make_loaders('mel', augment=True)
train_loader_mfcc, val_loader_mfcc, test_loader_mfcc = make_loaders('mfcc', augment=True)
train_loader_wave, val_loader_wave, test_loader_wave = make_loaders('wave', augment=True)
```

```python

def plot_wave_and_spec(path, n_mels=NUM_MELS):
    wav = load_wave(path)
    plt.figure(figsize=(12,3))
    plt.title('Waveform')
    plt.plot(wav.numpy())
    plt.show()

    spec = amplitude_to_db(mel_transform(wav)).clamp(min=-80.0)
    plt.figure(figsize=(12,4))
    plt.title('Log-mel Spectrogram')
    plt.imshow(spec.numpy(), aspect='auto', origin='lower')
    plt.colorbar()
    plt.show()

sample_path = train_df.iloc[0]['path']
plot_wave_and_spec(sample_path)
```

```python

class CNN2D(nn.Module):
    def __init__(self, in_ch=1, num_classes=8):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(in_ch, 32, 3, padding=1), nn.BatchNorm2d(32), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(), nn.MaxPool2d((2,2)),
            nn.Conv2d(128, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(), nn.AdaptiveAvgPool2d((1,1)),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(256, 128), nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        return self.classifier(x)

class CRNN(nn.Module):
    def __init__(self, in_ch=1, num_classes=8, hidden=128):
        super().__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(in_ch, 32, 3, padding=1), nn.ReLU(), nn.MaxPool2d((2,2)),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d((2,2)),
            nn.Conv2d(64, 128, 3, padding=1), nn.ReLU(), nn.MaxPool2d((2,2)),
        )
        self.rnn = nn.LSTM(input_size=128, hidden_size=hidden, num_layers=1, batch_first=True, bidirectional=True)
        self.head = nn.Sequential(
            nn.Linear(hidden*2, num_classes)
        )

    def forward(self, x):  # x: [B,1,F,T]
        feats = self.cnn(x)          # [B, C=128, F', T']
        B, C, Fp, Tp = feats.shape
        seq = feats.mean(dim=2)      # average over freq -> [B, C, T']
        seq = seq.permute(0,2,1)     # [B, T', C]
        out,_ = self.rnn(seq)        # [B, T', 2H]
        out = out.mean(dim=1)        # temporal pooling
        return self.head(out)

class CNN1D(nn.Module):
    def __init__(self, in_ch=1, num_classes=8):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_ch, 32, 9, stride=2, padding=4), nn.BatchNorm1d(32), nn.ReLU(),
            nn.Conv1d(32, 64, 9, stride=2, padding=4), nn.BatchNorm1d(64), nn.ReLU(),
            nn.Conv1d(64,128, 9, stride=2, padding=4), nn.BatchNorm1d(128), nn.ReLU(),
            nn.Conv1d(128,256, 9, stride=2, padding=4), nn.BatchNorm1d(256), nn.ReLU(),
            nn.AdaptiveAvgPool1d(1),
        )
        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):   # [B,1,T]
        z = self.net(x)
        return self.fc(z)
```

```python
def train_one_epoch(model, loader, optimizer, scaler, criterion, epoch, total_epochs):
    model.train()
    losses, preds, gts = [], [], []
    print(f"\n---- Epoch {epoch}/{total_epochs} ----")

    for batch_idx, (xb, yb, _) in enumerate(loader):
        xb = xb.to(DEVICE, non_blocking=True)
        yb = yb.to(DEVICE, non_blocking=True)

        optimizer.zero_grad(set_to_none=True)
        if scaler is not None:
            with torch.amp.autocast('cuda', enabled=(DEVICE.type=='cuda')):
                logits = model(xb)
                loss = criterion(logits, yb)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            logits = model(xb)
            loss = criterion(logits, yb)
            loss.backward()
            optimizer.step()

        losses.append(loss.item())
        preds.extend(logits.argmax(1).detach().cpu().tolist())
        gts.extend(yb.detach().cpu().tolist())

        if (batch_idx + 1) % 10 == 0:
            print(f"  Batch {batch_idx+1}/{len(loader)} | Loss: {loss.item():.4f}")

    acc = accuracy_score(gts, preds)
    f1 = f1_score(gts, preds, average='weighted', zero_division=0)
    return float(np.mean(losses)), acc, f1

@torch.no_grad()
def evaluate(model, loader, criterion):
    model.eval()
    losses, preds, gts = [], [], []
    for xb, yb, _ in loader:
        xb = xb.to(DEVICE, non_blocking=True)
        yb = yb.to(DEVICE, non_blocking=True)
        logits = model(xb)
        loss = criterion(logits, yb)
        losses.append(loss.item())
        preds.extend(logits.argmax(1).cpu().tolist())
        gts.extend(yb.cpu().tolist())
    acc = accuracy_score(gts, preds)
    f1 = f1_score(gts, preds, average='weighted', zero_division=0)
    return float(np.mean(losses)), acc, f1, np.array(gts), np.array(preds)

def fit_model(model, loaders, epochs=EPOCHS, lr=LR, wd=WEIGHT_DECAY, tag='model'):
    train_loader, val_loader, test_loader = loaders  # <-- USE what was passed in
    model = model.to(DEVICE)
    optimizer = AdamW(model.parameters(), lr=lr, weight_decay=wd)
    scheduler = CosineAnnealingLR(optimizer, T_max=epochs)
    criterion = nn.CrossEntropyLoss()
    scaler = torch.amp.GradScaler('cuda') if (AMP and DEVICE.type=='cuda') else None  # (same API, just confirming)

    history = {'train_loss':[], 'train_acc':[], 'train_f1':[],
               'val_loss':[],   'val_acc':[],   'val_f1':[]}

    best_val = -1.0
    best_path = f'best_{tag}.pt'
    patience, patience_ctr = 7, 0

    for ep in range(1, epochs+1):
        tr_loss, tr_acc, tr_f1 = train_one_epoch(model, train_loader, optimizer, scaler, criterion, ep, epochs)
        va_loss, va_acc, va_f1, _, _ = evaluate(model, val_loader, criterion)
        scheduler.step()

        history['train_loss'].append(tr_loss); history['train_acc'].append(tr_acc); history['train_f1'].append(tr_f1)
        history['val_loss'].append(va_loss);   history['val_acc'].append(va_acc);   history['val_f1'].append(va_f1)

        print(f"[EP {ep:02d}] train: loss={tr_loss:.4f} acc={tr_acc:.3f} f1={tr_f1:.3f} | val: loss={va_loss:.4f} acc={va_acc:.3f} f1={va_f1:.3f}")

        if va_f1 > best_val:
            best_val = va_f1
            torch.save({'state_dict': model.state_dict(), 'label_map': CLASSES}, best_path)
            patience_ctr = 0
        else:
            patience_ctr += 1
            if patience_ctr >= patience:
                print("Early stopping.")
                break

    print("  Validating...")
    ckpt = torch.load(best_path, map_location=DEVICE, weights_only=False)
    model.load_state_dict(ckpt['state_dict'])
    te_loss, te_acc, te_f1, y_true, y_pred = evaluate(model, test_loader, criterion)

    return model, history, {'loss':te_loss,'acc':te_acc,'f1':te_f1,'y_true':y_true,'y_pred':y_pred}
```

```python

import itertools

def plot_curves(history, title='Training Curves'):
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1); plt.title('Loss'); 
    plt.plot(history['train_loss'], label='train'); plt.plot(history['val_loss'], label='val'); plt.legend()
    plt.subplot(1,2,2); plt.title('F1'); 
    plt.plot(history['train_f1'], label='train'); plt.plot(history['val_f1'], label='val'); plt.legend()
    plt.suptitle(title); plt.show()

def plot_conf_mat(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred, labels=list(range(len(classes))))
    fig = plt.figure(figsize=(7,6))
    plt.imshow(cm, interpolation='nearest')
    plt.title('Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.show()

def plot_umap(embeds, labels, title='Embedding UMAP'):
    if not HAS_UMAP:
        print("umap-learn is not installed. Skipping UMAP plot."); return
    reducer = umap.UMAP(n_components=2, random_state=SEED)
    z = reducer.fit_transform(embeds)
    plt.figure(figsize=(7,6))
    for i, cls in enumerate(CLASSES):
        idx = (labels==i)
        plt.scatter(z[idx,0], z[idx,1], s=12, label=cls, alpha=0.7)
    plt.legend(); plt.title(title); plt.show()
```

## Train Model 1: **CNN2D on Log-Mel**

```python

cnn2d = CNN2D(in_ch=1, num_classes=NUM_CLASSES)
cnn2d, hist_cnn2d, test_cnn2d = fit_model(cnn2d, (train_loader_mel, val_loader_mel, test_loader_mel), tag='cnn2d_mel')
plot_curves(hist_cnn2d, title='CNN2D (mel)')
plot_conf_mat(test_cnn2d['y_true'], test_cnn2d['y_pred'], CLASSES)
print('Test metrics (CNN2D mel):', test_cnn2d)
```

## Train Model 2: **CRNN on Log-Mel**

```python

crnn = CRNN(in_ch=1, num_classes=NUM_CLASSES, hidden=128)
crnn, hist_crnn, test_crnn = fit_model(crnn, (train_loader_mel, val_loader_mel, test_loader_mel), tag='crnn_mel')
plot_curves(hist_crnn, title='CRNN (mel)')
plot_conf_mat(test_crnn['y_true'], test_crnn['y_pred'], CLASSES)
print('Test metrics (CRNN mel):', test_crnn)
```

## Train Model 3: **CNN1D on Waveform**

```python

cnn1d = CNN1D(in_ch=1, num_classes=NUM_CLASSES)
cnn1d, hist_cnn1d, test_cnn1d = fit_model(cnn1d, (train_loader_wave, val_loader_wave, test_loader_wave), tag='cnn1d_wave')
plot_curves(hist_cnn1d, title='CNN1D (wave)')
plot_conf_mat(test_cnn1d['y_true'], test_cnn1d['y_pred'], CLASSES)
print('Test metrics (CNN1D wave):', test_cnn1d)
```

## Embedding Visualization (UMAP) — from CRNN penultimate layer

```python

@torch.no_grad()
def extract_embeddings(model, loader):
    model.eval()
    feats_all, labels_all = [], []
    for xb, yb, _ in loader:
        xb = xb.to(DEVICE)
        # forward until before final linear for CRNN
        conv = model.cnn(xb)
        seq = conv.mean(dim=2).permute(0,2,1)
        out,_ = model.rnn(seq)
        pooled = out.mean(dim=1)  # [B, 2H]
        feats_all.append(pooled.cpu().numpy())
        labels_all.append(yb.numpy())
    return np.concatenate(feats_all, axis=0), np.concatenate(labels_all, axis=0)

emb, lab = extract_embeddings(crnn, test_loader_mel)
plot_umap(emb, lab, title='CRNN Embeddings (UMAP)')
```
